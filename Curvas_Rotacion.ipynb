## 1. Materia Oscura

### 1.A ¿Qué es la materia oscura?

Es muy probable que, incluso sin haber estudiado nada relacionado directamente con cosmología, en algún momento hayamos escuchado hablar de la "materia oscura" en algún medio de entretenimiento como videojuegos, series, comics, etc. Ya sea que haya sido de esta forma, por algún comentario en clase, alguna charla o vídeo divulgativo u otras fuentes seguramente lo primero que se nos vino a la cabeza fue la pregunta ¿Qué es la materia oscura? De forma breve la respuesta sería: no sabemos. Sin embargo, si quisiéramos ahondar un poco más, una respuesta más completa sería que, aunque no sabemos *¿Qué?* es, si sabemos *¿Cómo?* se comporta.

Desde que fue formulada por primera vez en 1933 por Fritz Zwicky [1], el tópico de la materia oscura es uno de los mayores retos de la cosmología y la astrofísica moderna [1]. En general lo entendemos como "una forma hipotética de materia que no emite suficiente radiación electromagnética como para ser detectada con los medios técnicos actuales" [2]. Esto quiere decir que la materia oscura no emite luz (de ahí su nombre), no absorbe ningún tipo de radiación de forma apreciable y no interactúa electromagéticamente de forma detectable [3]. Al resto de materia (la "común" pudiéramos decir) se le denomina "materia bariónica".
Uno pudiera pensar que un objeto que suena tan *exótico* en realidad sería raro, pero la realidad es que no. Según el modelo $\Lambda$CDM (el modelo cosmológico estándar) esta consituye cerca del 27% del contenido total de masa-energía del universo y al rededor del 85% de toda la materia [4, 5] **¡Eso es más de 5 veces la materia bariónica!** Por lo que podríamos decir que, a escala cosmológica, estamos rodeados por materia oscura. 

![Proporción de materia oscura](Imagenes/Planck_cosmic_recipe.jpg)

**Figura 1.** Proporción de materia oscura, materia bariónica y energía oscura en el universo.

**Fuente:** ESA (2013), *Planck Cosmic Recipe*.

Además de su abundancia y su falta de interacción electromagnética, también se sabe que es fría, es escencialmente no colisional, sus partículas interaccionan muy débilmente y debió existir desde los momentos iniciales del big bang aunque aún desconocemos su composición microscópica [1, 2]. Todas estas propiedades se obtienen a partir de la evidencia observacional de esta (véase la siguiente sección).

La importancia de entender la materia oscura radica en su relación con problemas como la composición del universo, la formación de las galaxias y otras estructuras cosmológicas, la teoría de la gravitación y la relatividad [2, 6, 7].

### 1.C Hipótesis Alternativas y Estado Actual

Aunque el concenso general actual es trabajar con el modelo de la materia oscura, existen hipótesis que buscan explicar las mediciones atribuidas a la materia oscura sin la necesidad de introducir una nueva forma de materia.

Normalmente estas hipótesis se centran en la idea de que nuestra comprensión de la gravedad está incompleta y no es completamente correcta a escalas galácticas o cosmológicas [15] o bien tratan de abordarla desde otra perspectiva, como que no se trata de una interacción fundamental sino de un fenómeno relacionado con información cuántica y entropía [16]. Aún así el problema de estas hipótesis es que no logra explicar todos los fenómenos asociados a la materia oscura de manera simultánea y más bien su éxito se concentra en puntos en concreto.

Es por eso que el modelo $\Lambda$CDM (que considera materia oscura fría y energía oscura) se considera el más completo y consistente para describir las observacionales cosmológicas actuales [14, 17]. Además la evidencia observacional que respalda que el 85% de la materia del universo debe tratarse de una no visible es abrumadora [2].

Aún así, existen aún muchas preguntas relacionadas con la materia oscura. La principal siendo que no sabemos cómo se compone, pero también hay problemáticas sobre cómo podríamos estudiarla o de qué forma se encuentra en los halos de materia oscura.

### 2.A Lo que esperábamos y lo que observamos

Supongamos que observamos una galaxia con forma en espiral y quisiéramos saber cómo es su velocidad en cada punto.
Lo primero que podría saltarnos a la mente es pensarlo en un vórtice (digamos como el agua girando en un desague). Si quisiéramos hacer lo análogo con un vórtice lo que haríamos sería revisar la conservación del momento angular [19]: 

$$
v_\theta (r) = \frac{\Gamma}{2 \pi r} 
$$
Donde:
* $v_\theta$ es la velocidad tangencial.
* r es la distancia al centro.
* $\Gamma$ es la circulación. En este caso, supongamos que es constante.

No es difícil ver que la velocidad decae a medida que más nos alejamos del centro (es decir, a medida que r crece) [19].

Viéndolo ahora con las leyes de Kepler y la mecánica newtoniana, para una estrella de masa m orbitando a una distancia r, la fuerza gravitatoria proporciona la fuerza centrípeta [18]:
$$
\frac{GM(r)m}{r^2} = \frac{mv^2}{r}
$$
Donde: 
* G es la constante gravitacional: $6.67430×10^{−11} m^3 kg^{−1} s^{−2}$
* M(R) es la masa de la galaxia encerrada en el radio "r".

De aquí se sigue:
$$
v(r) = \sqrt{\frac{GM(r)}{r}}
$$

Si consideramos un r suficientemente grande, que contenga prácticamente toda la masa de la galaxia, entonces [18]:

$$
v(r) \propto r^{-1/2}
$$

Es decir, entre más nos alejamos del centro la velocidad debería decaer. Sin embargo, como podemos apreciar en la **Figura 2**, esto no es lo que observamos. La velocidad se mantiene **casi constante** [2, 10, 18]. Esto sólo tiene sentido matemáticamente si la masa sigue creciendo a medida que aumenta la distancia:
$$
M(r) \propto r
$$
Lo que nos indicaba que había materia presente que no podíamos ver y eventualmente llevó a la idea de los halos de materia oscura [1].

Eston son una estructura extensa y aproximádamente esférica que rodea las galaxias y cúmulos de galaxias. Se extienden más allá del disco visible y en una galaxia espiral puede ser varias veces mayor que el radio observable de la galaxia [14,18]. Se les considera como componentes fundamentales para explicar la estabilidad dinámica y la formación de estructuras cósmicas [2].

### 2.C Aplicación de la estadística Bayesiana

Con la aparición de nuevos y más completos catálogos también surge el problema de tener que trabajar con una cantidad cada vez mayor de datos, lo cual se suele abordar desde la estadística. Sin embargo ¿Cómo podemos trabajarlos si tenemos las consecuencias pero no las causas? Thomas Bayes se hizo esta misma pregunta [29].

En sí, podríamos describir el problema central que estudió como "Si observo los resultados de un experimento ¿Qué puedo inferir acerca de la causa que los produjo?" [29]. Esto es un problema de inferencia inversa, algo que la probabilidad clásica no respondía.

Matemáticamente, partimos de la definición de probabilidad condicional:
$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

y también:

$$
P(B|A) = \frac{P(A \cap B)}{P(A)}
$$

Igualando ambas expresiones:
$$
P(A \cap B) = P(A|B)P(B) = P(B|A)P(A)
$$

Obtenemos:
$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$$

Lo interesante aquí no viene de la ecuación en sí mismo, sino en su interpretación [30]. Supongamos que:
* A: "La hipótesis es verdadera".
* B: "Hemos observado cierta evidencia".

Entonces:
* P(A): Creencia inicial.
* P(B|A): ¿Qué tan compatible es la evidencia con la hipótesis?
* P(A|B): Creencia actualizada despúes de observar la evidencia.

Luego el conocimiento previo debe modificarse de manera proporcional a qué tan bien explica la hipótesis los datos observados [30].

De esta forma, la estadística bayesiana es un método deductivo basado en el teorema de Bayes, que permite trabajar la incertidumbre sobre un parámetro físico mediante distribuciones de probabilidad [31]. Nos ayuda a trabajar los datos cuando su incertidumbre o correlación nos es desconocida y es una de las herramientas más poderosas de la cosmología de las últimas décadas. En nuestro caso, la ecuación que utilizaríamos sería [31]:

$$
P(\theta | D) = \frac{P(D|\theta)P(\theta)}{P(D)} 
$$

donde:
* D representa los datos observacionales.
* $\theta$ es el conjunto de parámetros del modelo.
* P($\theta$) es la distribución previa (prior).
* P(D|$\theta$) es la verosimilitud (likelihood).
* P($\theta$|D) es la distribución posterior (posterior).

Si pusiéramos en contexto de las curvas de rotación, los parámetros $\theta$ podrían incluir:
* Masa del halo.
* Radio de escala.
* Concentración del halo.
* Parámetros de modelos alternativos.

#### 2.C.a Métodos MCMC y Algoritmo Goodman & Weare

Por lo general en los problemas cosmológicos y de astrofísica la distribución posterior no puede obtenerse analíticamente, es por eso que se utilizan métodos *Markov Chain Monte Carlo* (MCMC) [32].

Las cadenas de Markov (Markov Chains) son procesos aleatorios que tienen la propiedad de que, tomando en cuenta un estado actual, el futuro es independiente del pasado [32]. En otras palabras, el estado futuro depende únicamente del estado actual y no de cómo se llegó a él [32].

Un ejemplo burdo sería si quisiéramos saber si el día de mañana puede o no llover en nuestra ciudad. Estaríamos de acuerdo que si el día de hoy está soleado, lo más seguro es que al día siguiente siga soleado o a lo mucho se nuble pero sería extraño que lloviera. Si queremos saber el clima al día siguiente no necesitamos saber cómo estuvo hace un mes, sólo importa el estado actual.

![Lloro-Colombia](Imagenes/Lloro_Colombia.jpg)

**Figura 12.** Lloró, Colombia. Considerada la ciudad más lluviosa del mundo [34].

**Accesible en:** https://www.colombiaturismoweb.com/DEPARTAMENTOS/CHOCO/MUNICIPIOS/LLORO/LLORO.htm .

Las cadenas de Markov nos resultan útiles ya que podemos trabajar nuestras mediciones incluso si no conocemos los estados iniciales.

Por su parte los métodos Monte Carlo son convenientes cuando las distribuciones de probabilidad de estos eventos se vuelven demasiado complicadas para resolverlas analíticamente o por medio de nuestras computadoras [32].

Los algoritmos MCMC generan una secuencia de muestras cuya distribución converge a la distribución posterior buscada [31], es decir que en lugar de calcular una distribución difícil mejor construimos una cadena de Markov que pase más tiempo en las regiones más probables. A partir de estas se pueden estimar valores más probables de los parámetros, intérvalos de credibilidad, correlaciones, incertidumbres del modelo, etc [31].

Ejemplo: Supongamos que estamos en una ciudad dividida en cuadras y queremos ir a dónde vive la mayor cantidad de gente pero no tenemos un mapa de densidad poblacional. Nustra única herramienta es caminar por la ciudad.

Si elegimos cuadras al azar, pasaríamos muchísimo tiempo visitando zonas que no son muy relevantes. Una estrategia alternativa sería estar un tiempo en una cuadra y luego elegir una cuadra vecina, si esta parece más poblada entonces vamos pero si parece menos poblada la dejamos para después. Así pasaríamos más tiempo en las regiones más densamente pobladas. La frecuencia con la que visitamos uan región termina reflejando su probabilidad.

![Edificio Altino Arantes](Imagenes/Webysther_20190306131041_-_Edifício_Altino_Arantes.jpg)

**Figura 12.** Edificio Altino Arantes en Sao Paulo capital, Brasil. La ciudad más poblada de América [35].

**Fuente:** Webysther Nunes (2019).

Los métodos MCMC son especialmente útiles en curvas de rotación porque los moderlos de los halos suelen involucrar parámetros altamente correlacionados [31].

Existen varios tipos de algoritmos MCMC, el más popular quizás sea el Metropolis-Hastings [32], sin embargo en este trabajo nos centraremos en el algoritmo Goodman & Weare debido a su practicidad a la hora de trabajar en python, R o Julia.

Introducido por Jonathan Goodman y Jonathan Weare en 2010 en su trabajo *"Ensemble samplers with affine invariance."* [33], el algoritmo Goodman & Weare (G&W) se diferencía de altoritmos tradicionales como Metropolis-Hastings ya que utiliza un conjunto de cadenas simultáneas denominadas *walkers*. Cada *walker* explora el espacio de parámetros utilizando la información de los demás *walkers* [33]. Esto le da la ventaja de que es *invariante afín* [33], es decir, matiene su eficiencia incluso si los parámetros presentan escalas diferentes o fuertes correlaciones.

El algoritmo G&W se popularizó en astronomía y cosmología gracias a la biblioteca EMCEE, desarrollada por Daniel Foreman-Mackey y colaboradoes [36]. Esta es una imprementación en Python, estable y bien probada, del muestrador para MCMC [36]. Otros lenguajes también tienen implementaciones que trabajan con G&W u otros algoritmos MCMC. En R encontramos *LaplaceDemon* [37] la cual incluye eplícitamente dicho procedimiento, además de *Bayesian Tools* [38] y *rstan* [39] que trabajan con otros mecanismos MCMC. Para Julia (lenguaje que se ha vuelto muy popular debido a su velocidad) el equivalente a EMCEE sería *AffinelnvariantMCMC.jl* [40] aunque también tenemos *Turin.jl* [41] que actualmente es el marco bayesiano más utilizado en Julia.

### 3.C El Modelo

Se refiere a la descripción matemática de nuestro problema. En este caso ya está definida en la literatura, así que partiremos de la relación física que se usa en curvas de rotación [45].

$$
V_{obs}^{2}(r) = V_{gas}^{2}(r) + V_{disk}^{2}(r) + V_{bul}^{2}(r) + V_{DM}^{2}(r)
$$

Para facilidad consideraremos:
$$
V_{DM} = V_{h} = cte.
$$

Sustituyendo esto en la ecucación, renombrando $V_{obs}$ como $V_{mod}$ y despejando para $V_{mod}$ se obtiene:

$$
V_{mod} = \sqrt{V_{gas}^{2}(r) + V_{disk}^{2}(r) + V_{bul}^{2}(r) + V_{h}^{2}}
$$

E intentaremos encontrar $V_h$ usando EMCEE. Cabe mencionar que este modelo es poco riguroso trabajarlo así ya que estamos fijando varios parámetros y trabajando únicamente sobre uno (que además estamos considerando constante). Aún así para fines académicos es útil para ilustrar cómo funciona un algoritmo de este tipo.

Si se quisiera hacer algo más completo se podría trabajar con la siguiente expresión:

$$
V_{obs}^{2}(r) = V_{gas}^{2}(r) + \Upsilon_d V_{disk}^{2}(r) + \Upsilon_b V_{bul}^{2}(r) + V_{DM}^{2}(r)
$$

Donde: 
* $\Upsilon_d$ es la razón masa-luz del disco.
* $\Upsilon_b$ es la razón masa-luz del bulbo.
* $V_DM$ es la contribución del halo de materia oscura.

Se podría trabajar con estos parámetros libres o incluso fijar dos de ellos (digamos, $\Upsilon_d$ y $\Upsilon_b$) pero aumenta un poco la complejidad del problema. Para mantenerlo como algo meramente ilustrativo de MCMC vamos a fijar toda la parte bariónica y trabajar sólo con el parámetro $V_h$.

Ahora definiremos nuestro modelo en Python.

```python
#Ejecuta esta celda para definir nuestro modelo como una función. Su nombre será "model_velocity".
#Depende de los parámetros vh, vgas, vdisk y vbul. En otras palabras, es la ecuación que usamos para V_mod.
def model_velocity(Vh,
                   Vgas,
                   Vdisk,
                   Vbul):

    return np.sqrt(
        Vgas**2 +
        Vdisk**2 +
        Vbul**2 +
        Vh**2
    )
```

También es importante definir nuestro vector de parámetros $\theta$. Como ya mencionamos anteriormente, sólo nos interesa un parámetro, así que:
$$
\theta = V_h
$$

```python
#Ejecuta esta celda.
ndim = 1
```

### 3.D El Prior

El prior (o distribución a priori) representa lo que sabemos o asumismos sobre los parámetros antes de incorporar los datos observacionales [46]. Matemáticamente se escribe:

$$
P(\theta)
$$

En nuestro caso debemos imponer que la velocidad sea no-negativa, pero debemos considerar un límite máximo ya que si lo dejamos abierto tendremos problemas más adelante con el likelihood.

Una buena regla sería asumir un valor máximo a partir de la velocidad máxima observada [46]. En este caso, de nuestro conjunto de datos [42], podemos ver que la velocidad observada más alta fue de 157 km/s. Así podríamos asumir un límite de 300 km/s (al rededor del doble).

$$
0 < V_h < 300
$$

Entonces:

$$
P(V_h)=
\begin{cases}
1, & 0 < V_h < 300 \\
0, & \text{otro caso}
\end{cases}
$$

Como EMCEE trabaja con logaritmos [36], nosotros trabajaremos con el logaritmo del prior para facilitar los cálculos.

```python
#Ejecuta esta celda, esta define nuestra función del logaritmo del prior "log_prior".
def log_prior(theta):

    Vh = theta[0]

    if 0 < Vh < 300:
        return 0.0

    return -np.inf
```

### 3.E Construir la función Chi-Cuadrada ($\chi^2$)

Para poder comprar la observación con el modelo hacemos uso de la función *Chi-cuadrada* ($\chi^2$) [46] mediante:

$$
\chi^2 = \sum_i{\frac{(v_{obs,i}-v_{mod,i})^2}{\sigma^2_i}}
$$

idealmente $v_{obs,i}=v_{mod,i}$, es decir que cada velocidad del modelo es exactamente igual a la observada. Sin embargo esto no sucedería a menos que tuviéramos un modelo 100% preciso (algo matemáticamente improbable).

```python
#Ejecuta esta celda para definir la función chi-cuadrada como "chi2"
def chi2(theta,
         Vobs,
         errV,
         Vgas,
         Vdisk,
         Vbul):

    Vh = theta[0]

    Vmodel = model_velocity(
        Vh,
        Vgas,
        Vdisk,
        Vbul
    )

    return np.sum(
        ((Vobs - Vmodel)/errV)**2
    )
```

### 3.F Likelihood

La likelihood (verosimilitud) mide que tan bien un conjunto de parámetros reproduce los datos observados [46]. Se define como:

$$
P(D|\theta)
$$

y se interpreta como "la probabilidad de observar los datos D suponiendo que los parámetros son $\theta$".

Se espera que el comportamiento de nuestros errores sea gaussiano [36] así que:

$$
\mathcal{L} \propto e^{-\chi^2/2}
$$

donde $\mathcal{L}$ es la likelihood. Es fácil ver que luego con el logaritmo se sigue:

$$
\ln \mathcal{L} = -\frac{\chi^2}{2}
$$

Ahora vamos con el código:

```python
#Ejecuta esta celda. Definimos el logaritmo del likelihood como log_likelihood.
def log_likelihood(theta,
                   Vobs,
                   errV,
                   Vgas,
                   Vdisk,
                   Vbul):

    return -0.5 * chi2(
        theta,
        Vobs,
        errV,
        Vgas,
        Vdisk,
        Vbul
    )
```

### 3.G La Posterior

La posterior es el resultado final del análisis bayesiano [46], combina la información previa (prior) con la información contenida en los datos (likelihood) y se obtiene a partir del teorema de Bayes:

$$
P(\theta | D) = \frac{P(D|\theta)P(\theta)}{D}
$$

Donde:
* $P(\theta|D)$ es la posterior.
* $P(D|\theta)$ es la likelihood.
* $P(\theta)$ es la prior.
* $P(D)$ es la evidencia bayesiana.

Pero como EMCEE no necesita calcular la evidencia [36], normalmente se usa:

$$
P(\theta|D) \propto P(D|\theta) P(\theta)
$$

Ahora bien, al trabajar con logaritmos también es más sencilla esta parte, ya que si aplicamos el logaritmo a ambas partes de la igualdad obtenemos:

$$
\ln(P(\theta|D)) \propto \ln (P(D|\theta) P(\theta)) = \ln (P(D|\theta)) + \ln (P(\theta))
$$

Y como recordaremos $\ln (P(D|\theta))$ es el logaritmo del likelihood y $\ln (P(\theta))$ el del prior. Ahí está otra ventaja de trabajar sobre los logaritmos en vez de sobre las funciones en sí.

```python
#Ejecuta esta celda para definir el logaritmo del posterior. Lo llamaremos log_probability.
def log_probability(theta,
                    Vobs,
                    errV,
                    Vgas,
                    Vdisk,
                    Vbul):

    lp = log_prior(theta)

    if not np.isfinite(lp):
        return -np.inf

    return (
        lp +
        log_likelihood(
            theta,
            Vobs,
            errV,
            Vgas,
            Vdisk,
            Vbul
        )
    )
```

### 3.H Walkers y EMCEE

Lo último que nos faltaría para poder ejecutar EMCEE es inicializar a nuestros *walkers*, que como ya dijimos nos ayudarán a explorar el espacio de parámetros [36]. Para esto hay que definir la cantidad de walkers y su punto de partida.

Para la cantidad de walkers, lo que se sugiere [36] es utilizar 2 walkers por cada parámetro del vector de parámetros. Sin embargo, ya que en nuestro caso es 1, tomaremos un número más grande. Es importante mencionar que entre más walkers se utilice menor diferencia hay (en tanto superen la proporción de 2 a 1) [36] y sólo hará más lento el proceso. En este caso consideraremos usar 32 walkers, aunque siéntete libre de modificar el código para aumentar o disminuir walkers y ver qué sucede.

En cuanto al punto inicial, lo que requerimos es que inicie dentro del espacio de parámetros [36]. En este caso, vamos a colocarlo en 50 km/s aunque igualmente puedes modificar este punto inicial y ver qué pasa.

```python
#Ejecuta esta celda para inicializar los walkers.

#Cambia este valor para aumentar o disminuir el número de walkers. Por defecto usamos 32 en este ejemplo.
nwalkers = 32


# Cambia este valor para modificar el punto de inicio. Considera que debe ser no negativo y menor a 300. Por defecto usamos 50 en este ejemplo.
initial = 50

pos = (
    initial
    + 1e-2*np.random.randn(
        nwalkers,
        1
    )
)
```

Ahora si, con todo listo vamos a ejecutar EMCEE.

```python
#Ejecuta esta celda para iniciar EMCEE. Puedes ejecutarla varias veces y ver cómo cambia el resultado.
#Tras ejecutar la celda, espera un poco a que aparezca el mensaje "✓ Muestreo completado".

sampler = emcee.EnsembleSampler(
    nwalkers,
    ndim,
    log_probability,
    args=(
        Vobs,
        errV,
        Vgas,
        Vdisk,
        Vbul
    )
)

#Si quieres ver todos los datos que arroja y el porcentaje de avance, puedes cambiar el "False" luego de "progress" por "True" y ejecutar
#la celda.
#Si lo haces, ignora los mensajes de error y espera a que termine. Si luego de haberlo hecho quieres que no se vean los datos, vuelve a 
#escribir "False" y ejecuta la celda.

sampler.run_mcmc(
    pos,
    5000,
    progress=False
)

print("✓ Muestreo completado")
```

### 3.I Burn-in y Resultados

Antes de pasar a revisar los resultados, tenemos que eliminar el Burn-In.

En las primeras etapas del MCMC, los walkers aún no alcanzan la distribución estacionaria (la más probable). Esta etapa se conoce como el Burn-in y por lo general debe ser ignorada [36].

Para indicar que la ignore, vamos a ejecutar la siguiente celda:

```python
#Ejecuta esta celda para limpair el Burn-In.

samples = sampler.get_chain(
    discard=1000,
    flat=True
)
```

Ahora si, ya podemos mostrar el parámetro estimado.

Este se obtiene usando percentiles de la distribución posterior, típicamente los percentiles 16 (aproximadamente a -$1\sigma$), 50 (la mediana) y 84 (aproximadamente a +$1\sigma$), correspondientes al valor central y a las incertidumbres de $1\sigma$ [46].

```python
#Ejecuta esta celda para obtener el resultado.

Vh_median = np.percentile(
    samples[:,0],
    50
)

Vh_low = np.percentile(
    samples[:,0],
    16
)

Vh_high = np.percentile(
    samples[:,0],
    84
)

print(f"Vh = {Vh_median:.2f} km/s")
print(f"Percentil 16 = {Vh_low:.2f} km/s")
print(f"Percentil 84 = {Vh_high:.2f} km/s")
```

Ahora bien ¿Este ajuste es bueno? Para esto vamos a compararlo con lo que el modelo nos indica.

Primero debemos definir nuestra función que será el "modelo".

```python
#Ejecuta esta celda, nos dará el modelo.
Vmodel = np.sqrt(
    Vgas**2 +
    Vdisk**2 +
    Vbul**2 +
    Vh_median**2
)
```

Luego calculamos el mejor ajuste de $\chi^2$[46]:

```python
#Ejecuta esta celda, comparará las v_obs con las v_model.
chi2_best = np.sum(
    ((Vobs - Vmodel)/errV)**2
)

print(f"Chi² = {chi2_best:.3f}")
```

Calculamos el $\chi^2$ reducido utilizando [46]:
$$
\chi^2_{\nu} = \frac{\chi^2}{N-p}
$$

donde:
* N es el número de observaciones.
* p es el número de parámetros libres.


```python
#Ejecuta esta celda. Calcula la chi cuadrada reducida.
N = len(Vobs)
p = 1

chi2_red = chi2_best/(N-p)

print(f"Chi² reducido = {chi2_red:.3f}")
```

Por lo general lo que querríamos sería un valor lo más cercano a 1 posible, ya que esto indicaría un buen ajuste. Si el resultado es mucho menor a 1 entonces los errores están sobreestimados, mientras que si es mayor a 2 el modelo es ineficiente [46].

Ahora bien, en nuestro caso es muchísimo mayor a 2 pero es de esperarse. Si recordamos, decidimos sobresimplificarlo para fines del ejercicio y es por eso que nuestro modelo no representa muy bien la realidad.

Podemos también revisar el parámetro estimado y el rango de velocidades:

```python
#Ejecuta esta celda, veremos los 3 parámetros: el estimado, el mínimo de nuestro conjunto de datos y el máximo.
print("Valor estimado =", Vh_median)
print("Valor mínimo =", np.min(Vobs))
print("Valor máximo =", np.max(Vobs))
```

Vemos que de hecho el valor estimado está dentro del rango y cerca del punto medio del rango de datos, esto nos dice que nuestra estimación no fue tan exagerada pero el valor de $\chi^2_\nu$ nos indica que (aunque probablmenete nuestro valor sea aceptable) la manera en la que describimos el problema se aleja mucho de la realidad (lo cual, como ya habíamos dicho, es comprensible).

### 3.J Reutilizar EMCEE

Luego de todo esto uno podría preguntarse "¿Entonces cada vez que quiera usar EMCEE en el mismo notebook tengo que definirlo una y otra vez?" Y en términos generales la respuesta es no. Siempre que el problema que abordas sea similar, puedes seguir utilizando todas las funciones que ya definimos y únicamente centrarse en importar los datos de manera correcta. Eso sí: Si se quiere que no se sobre-escriban los valores ya calculados o importados, hay que modificar algunas etiquetas.

Para mostrar cómo, ahora haremos EMCEE para la segunda galaxia que mencionamos (DDO 161).

Comenzamos importando los datos.

```python
#Ejecuta esta celda para importar nuestros datos como "data2".
data2 = np.loadtxt(
    "Data/DDO161_rotmod.dat",
    comments="#"
)

#Luego esta parte extrae las columnas y define la serie de datos. Nótese que ahora debemos de tener
#cuidado de poner "2" después de cada dato y especificar que es el set de datos "data2" para no
#sobreescribir los datos que teníamos anteriormente y que se importen los de la segunda galaxia.
Rad2    = data2[:,0]
Vobs2   = data2[:,1]
errV2   = data2[:,2]
Vgas2   = data2[:,3]
Vdisk2  = data2[:,4]
Vbul2   = data2[:,5]
SBdisk2 = data2[:,6]
SBbul2  = data2[:,7]
```

Como tanto el modelo como los priors, likelihood, posteriors y $\chi^2$ son las mismas, entonces ya podríamos ejecutar los walkers y el EMCEE.

Ahora bien, si queremos hacerlo reutilizable conviene hacer un pequeño cambio y en vez de definir cada parte del proceso por separado ponerlo todo junto en una misma función. Esto no sólo nos ahorra espacio de trabajo sino que ahora cada vez que queramos usarlo solo tenemos que llamar a la función.

Así:

```python
#Ejecuta esta celda para definir nuestro proceso EMCEE como run_EMCEE.
#Vemos que ahora el número de walkers, de pasos, punto de inicio y el burn-in lo definimos como
#variables de la función.
def run_emcee(
    Vobs,
    errV,
    Vgas,
    Vdisk,
    Vbul,
    nwalkers=32,
    nsteps=5000,
    burnin=1000,
    initial=50
):
#Aquí vemos la dimensión de nuestro vector de parámetros.
    ndim = 1

#Posiciones iniciales de los walkers. Tiene una pequeña perturbación aleatoria para evitar que
#ocupen el mismo punto. 
    pos = (
        initial
        + 1e-2*np.random.randn(
            nwalkers,
            ndim
        )
    )
#Creación del sampler (muestreador), aquí es donde el número de walkers entra.
    sampler = emcee.EnsembleSampler(
        nwalkers,
        ndim,
        log_probability,
        args=(
            Vobs,
            errV,
            Vgas,
            Vdisk,
            Vbul
        )
    )
#Se ejecuta el algorimo MCMC.
    sampler.run_mcmc(
        pos,
        nsteps,
        progress=False
    )
#Se descarta el Burn-In y se obtienen las muestras.
    samples = sampler.get_chain(
        discard=burnin,
        flat=True
    )
#Se estima Vh.
    Vh_median = np.percentile(
        samples[:,0],
        50
    )
# Límite inferior del intervalo de credibilidad (percentil 16).
    Vh_low = np.percentile(
        samples[:,0],
        16
    )
# Límite superior del intervalo de credibilidad (percentil 84).
    Vh_high = np.percentile(
        samples[:,0],
        84
    )

    return {
        "sampler": sampler,
        "samples": samples,
        "Vh": Vh_median,
        "Vh_low": Vh_low,
        "Vh_high": Vh_high
    }
```

Y ahora si queremos ejecutarlo, sólo falta ejecutar esta función con los datos de la galaxia que nos interesa.

```python
#Ejecuta esta celda para encontrar Vh para la segunda galaxia.
resultado2 = run_emcee(
    Vobs2,
    errV2,
    Vgas2,
    Vdisk2,
    Vbul2
)

print(resultado2["Vh"])
```

Podemos ejecutarlo también para la primera nuevamente (el resultado cambiará un poco por la aleatoriedad pero estará dentro del rango de los percentiles que obtuvimos anteriormente.

```python
#Ejecuta esta celda para encontrar Vh para la primer galaxia.
resultado1 = run_emcee(
    Vobs,
    errV,
    Vgas,
    Vdisk,
    Vbul
)

print(resultado1["Vh"])
```

Así podríamos obtener valores para distintas galaxias simplemente importando los datos y ejecutando la función que definimos.
